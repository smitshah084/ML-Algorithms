```python
X_df = pd.DataFrame(iris_data['data'])
X_df.columns = ['sepal_l','sepal_w','petal_l','petal_w']
y_df = pd.DataFrame(iris_data['target'])
y_df.columns = ['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df,y_df,random_state=104,test_size=0.25,shuffle=True) 
X_train, X_test, y_train, y_test = X_train.reset_index(), X_test.reset_index(), y_train.reset_index(), y_test.reset_index()
X_train, X_test, y_train, y_test = X_train.drop(['index'],axis=1), X_test.drop(['index'],axis=1), y_train.drop(['index'],axis=1), y_test.drop(['index'],axis=1)
```

```python
X_train = X_train.reindex()
X_train
```

```python
def sigmoid(X):
    return 1/(1+np.exp(X))
```

```python
def h(X,theta):
#     X:(m,n),theta:(n,1),h:(m,1)
    return sigmoid(X@theta)
```

```python
def j(X,y,theta):
    
    m = len(y)  
    h1 = sigmoid(h(X, theta))  
    cost = (-1/m) * np.sum(y * np.log(h1) + (1 - y) * np.log(1 - h1))
    
    return cost
```

```python
def G(X,y,theta,alpha):
    
    m = len(y)
    c = alpha/m
    temp_0 = X.transpose() # (n,m)
    temp_1 = h(X,theta) - y # (m,1)
    temp_2 = temp_0 @ temp_1 # (n,1)
    temp_3 = c * temp_2
    
    return theta - temp_3
```

```python
def normalise(X):
    
    X = (X - X.mean()) / X.std()
    
    return X
```

```python
def GD(X,y,alpha,thresh):
    
    X = normalise(X)
    X = X.values
    y = y.values 
    
    m = X.shape[0]
    n = X.shape[1]
    theta = np.random.rand(n,1)
    costs = [j(X,y,theta)]
    cost_diff = costs[0]
    iter_ = 0
    
#     while iter_ <= 1000:
    while cost_diff >= thresh:
        
        theta = G(X,y,theta,alpha)
        C = j(X,y,theta)
        cost_diff = costs[-1] - C
        costs.append(C)
        iter_ += 1
        
    plt.plot(costs)
    
    return theta
```

```python
alpha = 1e-3
thresh = 0.000003

theta = GD(X_train,y_train,alpha,thresh)
```
